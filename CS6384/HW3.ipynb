```python
import matplotlib.pyplot as plt
img = plt.imread("./data/pic.jpeg")
```

```python
plt.imshow(img)
```

```python
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]

#Standard formula to convert RGB image to grayscale image.
gray_img = 0.2989 * R + 0.5870 * G + 0.1140 * B
```

```python
plt.imshow(gray_img, cmap = 'gray')
```

```python
neg_img = 255 - gray_img
```

```python
plt.imshow(neg_img, cmap = 'gray')
```

```python
import matplotlib.image as mpimg
mpimg.imsave("neg_gray_img.jpeg", neg_img, cmap='gray')
```

```python
#Importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
img_name = 'data/pic.jpeg'
img = cv2.imread(img_name, cv2.IMREAD_UNCHANGED);
if(img is None) :
    print(sys.argv[0], "Failed to read image from ", img_name)
    sys.exit()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
rank = len(img.shape)
if(rank == 2) :
    gray_image = img
elif(rank == 3) :
    gray_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
else :
    print(sys.argv[0], "Can't handle unusual image ", img_name)
    sys.exit()
```

```python
plt.imshow(gray_image, cmap='gray')
```

```python
neg_img = cv2.bitwise_not(gray_image)
```

```python
plt.imshow(neg_img, cmap='gray')
```

```python
cv2.imwrite("data/neg_gray_img.jpeg", neg_img);
```
